### Conversion of data types  
**OBJECT TO FLOAT:**
- discounted_price
- actual_price
- rating

**OBJECT TO INT64:**
- rating_count 

```python
# Remove currency symbols and commas, then convert
df['discounted_price'] = pd.to_numeric(df['discounted_price'].str.replace('₹', '', regex=False).str.replace(',', '', regex=False), errors='coerce')
df['actual_price'] = pd.to_numeric(df['actual_price'].str.replace('₹', '', regex=False).str.replace(',', '', regex=False), errors='coerce')

# Clean and convert rating_count
df['rating_count'] = pd.to_numeric(df['rating_count'].str.replace(',', '', regex=False), errors='coerce')

# Convert ratings
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Confirm changes
df.dtypes
```

#### 4.3 Missing Value Treatment — Phase 1 Approach

- **Detected:** 2 NaN values in `rating_count` (numeric).
- **Chosen Treatment:** Median imputation (single pass).
- **Rationale:** Low proportion (0.136%) and distribution approximately normal. Fill with median (more robust than mean for skewed data like ratings count).
- **Note for Phase 2:** If downstream analysis shows sensitivity to this column’s values, revisit imputation method (possible median or predictive model imputation).


```python
df.fillna(df["rating_count"].median(), inplace =True) 
df.isna().sum()
```
